# Variable Aleatoria

## DistribuciÃ³n Uniforme

### **Problema1**

Ejercicio: El tiempo que espera un viajero en una estaciÃ³n de Bus se distribuye uniformemente entre 2 y 28.5 minutos  U(2;28,5). Â¿cuÃ¡l es la probabilidad de que un viajero tenga que esperar mÃ¡s de 5.28 minutos por su Bus?

a=2
b=28.5

image.png

```python
import matplotlib.pyplot as plt
import numpy as np

# ParÃ¡metros de la distribuciÃ³n
a = 2          # LÃ­mite inferior
b = 28.5       # LÃ­mite superior
x_val = 5.28   # Valor desde donde se calcula P(X > x_val)

# ------------------------
# CÃLCULO DE PROBABILIDAD
# ------------------------
# FÃ³rmula: P(X > x) = (b - x) / (b - a)
probabilidad = (b - x_val) / (b - a)
print(f"La probabilidad de que el viajero espere mÃ¡s de {x_val} minutos es aproximadamente: {probabilidad:.4f} ({probabilidad*100:.2f}%)")

# ------------------------
# GRÃFICO
# ------------------------
# Altura de la funciÃ³n de densidad
height = 1 / (b - a)

# Crear valores de x para graficar
x = np.linspace(a - 1, b + 1, 500)
y = np.where((x >= a) & (x <= b), height, 0)

# Crear figura
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='DistribuciÃ³n U(2, 28.5)', color='blue')

# Sombrear regiÃ³n P(X > x_val)
plt.fill_between(x, y, where=(x >= x_val) & (x <= b), color='orange', alpha=0.5, label=f'P(X > {x_val})')

# LÃ­nea vertical en x_val
plt.axvline(x_val, color='black', linestyle='--', label=f'x = {x_val}')

# Etiquetas y detalles
plt.title('DistribuciÃ³n Uniforme Continua: Espera de un viajero')
plt.xlabel('Minutos de espera')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)

# Mostrar grÃ¡fico
plt.show()
```

### **Problema 2:**

Dados los siguientes nÃºmeros aleatorios calcular variables aleatorias de una distribuciÃ³n U(6; 16)  
0,12
0,01
0,23
0,28
0,89

Dados los siguientes nÃºmeros aleatorios:

0.12, 0.01,Â 0.23,Â 0.28,Â 0.89

Calcular las variables aleatorias ð‘‹ âˆ¼ ð‘ˆ (6,16)correspondientes.

image.png

image.png

```python
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: NÃºmeros aleatorios dados (U[0,1])
numeros_aleatorios = np.array([0.12, 0.01, 0.23, 0.28, 0.89])

# Paso 2: ParÃ¡metros de la distribuciÃ³n uniforme U(6, 16)
a = 6
b = 16

# Paso 3: TransformaciÃ³n a U(6, 16)
variables_uniformes = a + (b - a) * numeros_aleatorios

# Mostrar resultados
for i, (r, x) in enumerate(zip(numeros_aleatorios, variables_uniformes), start=1):
    print(f"{i}. R = {r} â†’ X = {x:.2f}")

# Paso 4: GrÃ¡fica de la distribuciÃ³n U(6, 16)
x_vals = np.linspace(a - 1, b + 1, 500)
y_vals = np.where((x_vals >= a) & (x_vals <= b), 1 / (b - a), 0)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='DistribuciÃ³n U(6,16)', color='blue')

# Marcar los puntos transformados
plt.scatter(variables_uniformes, [1 / (b - a)] * len(variables_uniformes),
            color='red', zorder=5, label='Variables transformadas')

# Detalles del grÃ¡fico
plt.title('TransformaciÃ³n de nÃºmeros aleatorios a U(6,16)')
plt.xlabel('Valor de X')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

## DistribuciÃ³n Exponencial

### Problema 1

Ejercicio: Dados los siguientes nÃºmeros aleatorios calcular variables aleatorias de una distribuciÃ³n Exp(0,05)
0,12
0,01
0,23
0,28
0,89

image.png

```python
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: NÃºmeros aleatorios dados (U[0,1])
numeros_aleatorios = np.array([0.12, 0.01, 0.23, 0.28, 0.89])

# Paso 2: ParÃ¡metro de la distribuciÃ³n exponencial
lmbda = 0.05

# Paso 3: TransformaciÃ³n a Exp(0.05)
variables_exponenciales = -1 / lmbda * np.log(1 - numeros_aleatorios)

# Mostrar resultados
print("TransformaciÃ³n R â†’ X para Exp(0.05):")
for i, (r, x) in enumerate(zip(numeros_aleatorios, variables_exponenciales), start=1):
    print(f"{i}. R = {r} â†’ X = {x:.2f} minutos")

# Paso 4: GrÃ¡fica
x_vals = np.linspace(0, max(variables_exponenciales) + 20, 500)
pdf_vals = lmbda * np.exp(-lmbda * x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, pdf_vals, label='DistribuciÃ³n Exp(0.05)', color='blue')

# Marcar los valores transformados
plt.scatter(variables_exponenciales, lmbda * np.exp(-lmbda * variables_exponenciales),
            color='red', zorder=5, label='Variables transformadas')

# Detalles del grÃ¡fico
plt.title('TransformaciÃ³n de nÃºmeros aleatorios a distribuciÃ³n Exponencial(0.05)')
plt.xlabel('Tiempo (minutos)')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

## DistribuciÃ³n Triangular

### Problema1

Ejercicio: Dados los siguientes nÃºmeros aleatorios calcular variables aleatorias de una distribuciÃ³n Triangular(3, 8, 14)
0,12
0,01
0,23
0,28
0,89

image.png



```python
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: ParÃ¡metros de la distribuciÃ³n Triangular
a = 3   # mÃ­nimo
c = 8   # moda
b = 14  # mÃ¡ximo

# Paso 2: NÃºmeros aleatorios dados (R ~ U[0,1])
numeros_aleatorios = np.array([0.12, 0.01, 0.23, 0.28, 0.89])

# Punto de divisiÃ³n entre las dos fÃ³rmulas
F_c = (c - a) / (b - a)

# Paso 3: Transformar a Triangular(a, c, b)
valores_transformados = []
for r in numeros_aleatorios:
    if r < F_c:
        x = a + np.sqrt(r * (b - a) * (c - a))
    else:
        x = b - np.sqrt((1 - r) * (b - a) * (b - c))
    valores_transformados.append(x)

valores_transformados = np.array(valores_transformados)

# Mostrar resultados
print("TransformaciÃ³n R â†’ X para Triangular(3, 8, 14):")
for i, (r, x) in enumerate(zip(numeros_aleatorios, valores_transformados), start=1):
    print(f"{i}. R = {r} â†’ X = {x:.2f}")

# Paso 4: Graficar la distribuciÃ³n teÃ³rica y los valores transformados
x_vals = np.linspace(a, b, 500)
pdf_vals = np.where(
    x_vals < c,
    2 * (x_vals - a) / ((b - a) * (c - a)),
    2 * (b - x_vals) / ((b - a) * (b - c))
)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, pdf_vals, color='blue', label='DistribuciÃ³n Triangular(3,8,14)')

# Marcar los valores transformados
plt.scatter(valores_transformados,
            np.where(valores_transformados < c,
                     2 * (valores_transformados - a) / ((b - a) * (c - a)),
                     2 * (b - valores_transformados) / ((b - a) * (b - c))),
            color='red', zorder=5, label='Valores generados')

# Detalles del grÃ¡fico
plt.title('TransformaciÃ³n a distribuciÃ³n Triangular(3,8,14)')
plt.xlabel('Valor de X')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```
