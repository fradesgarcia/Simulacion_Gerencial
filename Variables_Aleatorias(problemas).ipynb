# Variable Aleatoria

## Distribución Uniforme

### **Problema1**

Ejercicio: El tiempo que espera un viajero en una estación de Bus se distribuye uniformemente entre 2 y 28.5 minutos  U(2;28,5). ¿cuál es la probabilidad de que un viajero tenga que esperar más de 5.28 minutos por su Bus?

a=2
b=28.5

image.png

```python
import matplotlib.pyplot as plt
import numpy as np

# Parámetros de la distribución
a = 2          # Límite inferior
b = 28.5       # Límite superior
x_val = 5.28   # Valor desde donde se calcula P(X > x_val)

# ------------------------
# CÁLCULO DE PROBABILIDAD
# ------------------------
# Fórmula: P(X > x) = (b - x) / (b - a)
probabilidad = (b - x_val) / (b - a)
print(f"La probabilidad de que el viajero espere más de {x_val} minutos es aproximadamente: {probabilidad:.4f} ({probabilidad*100:.2f}%)")

# ------------------------
# GRÁFICO
# ------------------------
# Altura de la función de densidad
height = 1 / (b - a)

# Crear valores de x para graficar
x = np.linspace(a - 1, b + 1, 500)
y = np.where((x >= a) & (x <= b), height, 0)

# Crear figura
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Distribución U(2, 28.5)', color='blue')

# Sombrear región P(X > x_val)
plt.fill_between(x, y, where=(x >= x_val) & (x <= b), color='orange', alpha=0.5, label=f'P(X > {x_val})')

# Línea vertical en x_val
plt.axvline(x_val, color='black', linestyle='--', label=f'x = {x_val}')

# Etiquetas y detalles
plt.title('Distribución Uniforme Continua: Espera de un viajero')
plt.xlabel('Minutos de espera')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)

# Mostrar gráfico
plt.show()
```

### **Problema 2:**

Dados los siguientes números aleatorios calcular variables aleatorias de una distribución U(6; 16)  
0,12
0,01
0,23
0,28
0,89

Dados los siguientes números aleatorios:

0.12, 0.01, 0.23, 0.28, 0.89

Calcular las variables aleatorias 𝑋 ∼ 𝑈 (6,16)correspondientes.

image.png

image.png

```python
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: Números aleatorios dados (U[0,1])
numeros_aleatorios = np.array([0.12, 0.01, 0.23, 0.28, 0.89])

# Paso 2: Parámetros de la distribución uniforme U(6, 16)
a = 6
b = 16

# Paso 3: Transformación a U(6, 16)
variables_uniformes = a + (b - a) * numeros_aleatorios

# Mostrar resultados
for i, (r, x) in enumerate(zip(numeros_aleatorios, variables_uniformes), start=1):
    print(f"{i}. R = {r} → X = {x:.2f}")

# Paso 4: Gráfica de la distribución U(6, 16)
x_vals = np.linspace(a - 1, b + 1, 500)
y_vals = np.where((x_vals >= a) & (x_vals <= b), 1 / (b - a), 0)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='Distribución U(6,16)', color='blue')

# Marcar los puntos transformados
plt.scatter(variables_uniformes, [1 / (b - a)] * len(variables_uniformes),
            color='red', zorder=5, label='Variables transformadas')

# Detalles del gráfico
plt.title('Transformación de números aleatorios a U(6,16)')
plt.xlabel('Valor de X')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

## Distribución Exponencial

### Problema 1

Ejercicio: Dados los siguientes números aleatorios calcular variables aleatorias de una distribución Exp(0,05)
0,12
0,01
0,23
0,28
0,89

image.png

```python
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: Números aleatorios dados (U[0,1])
numeros_aleatorios = np.array([0.12, 0.01, 0.23, 0.28, 0.89])

# Paso 2: Parámetro de la distribución exponencial
lmbda = 0.05

# Paso 3: Transformación a Exp(0.05)
variables_exponenciales = -1 / lmbda * np.log(1 - numeros_aleatorios)

# Mostrar resultados
print("Transformación R → X para Exp(0.05):")
for i, (r, x) in enumerate(zip(numeros_aleatorios, variables_exponenciales), start=1):
    print(f"{i}. R = {r} → X = {x:.2f} minutos")

# Paso 4: Gráfica
x_vals = np.linspace(0, max(variables_exponenciales) + 20, 500)
pdf_vals = lmbda * np.exp(-lmbda * x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, pdf_vals, label='Distribución Exp(0.05)', color='blue')

# Marcar los valores transformados
plt.scatter(variables_exponenciales, lmbda * np.exp(-lmbda * variables_exponenciales),
            color='red', zorder=5, label='Variables transformadas')

# Detalles del gráfico
plt.title('Transformación de números aleatorios a distribución Exponencial(0.05)')
plt.xlabel('Tiempo (minutos)')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

## Distribución Triangular

### Problema1

Ejercicio: Dados los siguientes números aleatorios calcular variables aleatorias de una distribución Triangular(3, 8, 14)
0,12
0,01
0,23
0,28
0,89

image.png



```python
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: Parámetros de la distribución Triangular
a = 3   # mínimo
c = 8   # moda
b = 14  # máximo

# Paso 2: Números aleatorios dados (R ~ U[0,1])
numeros_aleatorios = np.array([0.12, 0.01, 0.23, 0.28, 0.89])

# Punto de división entre las dos fórmulas
F_c = (c - a) / (b - a)

# Paso 3: Transformar a Triangular(a, c, b)
valores_transformados = []
for r in numeros_aleatorios:
    if r < F_c:
        x = a + np.sqrt(r * (b - a) * (c - a))
    else:
        x = b - np.sqrt((1 - r) * (b - a) * (b - c))
    valores_transformados.append(x)

valores_transformados = np.array(valores_transformados)

# Mostrar resultados
print("Transformación R → X para Triangular(3, 8, 14):")
for i, (r, x) in enumerate(zip(numeros_aleatorios, valores_transformados), start=1):
    print(f"{i}. R = {r} → X = {x:.2f}")

# Paso 4: Graficar la distribución teórica y los valores transformados
x_vals = np.linspace(a, b, 500)
pdf_vals = np.where(
    x_vals < c,
    2 * (x_vals - a) / ((b - a) * (c - a)),
    2 * (b - x_vals) / ((b - a) * (b - c))
)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, pdf_vals, color='blue', label='Distribución Triangular(3,8,14)')

# Marcar los valores transformados
plt.scatter(valores_transformados,
            np.where(valores_transformados < c,
                     2 * (valores_transformados - a) / ((b - a) * (c - a)),
                     2 * (b - valores_transformados) / ((b - a) * (b - c))),
            color='red', zorder=5, label='Valores generados')

# Detalles del gráfico
plt.title('Transformación a distribución Triangular(3,8,14)')
plt.xlabel('Valor de X')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```
